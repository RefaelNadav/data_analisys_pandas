```python
df_exchange = pd.read_csv('./data/USD_ILS Historical Data.csv')
```

```python
df_exchange['Date'] = pd.to_datetime(df_exchange['Date'], errors='coerce')
df_exchange
```

```python
df_us = pd.read_csv('./data/yield-curve-rates-1990-2023.csv')
df_us
```

```python
df_us['Date'] = df_us['Date'].str.strip()
df_us['Date'] = pd.to_datetime(df_us['Date'], format='%m/%d/%y', errors='coerce')
df_us['year'] = df_us['Date'].dt.year
df_us['month'] = df_us['Date'].dt.month
df_us.drop_duplicates(subset=['year', 'month'], inplace=True)
df_us["current_rate"] = df_us['1 Mo']
df_us["change_from_last_month"] = df_us["current_rate"].diff()
df_us = df_us[['Date', 'year', 'month', 'current_rate', 'change_from_last_month']]
df_us
```

```python
df_il = pd.read_csv('./data/interest_rate_il.csv')
df_il = df_il.iloc[3: 330]
df_il.tail()
```

```python
dct_months = dict(Jan=1, Feb=2, Mar=3, Apr=4, May=5, Jun=6, Jul=7, Aug=8, Sep=9, Oct=10, Nov=11, Dec=12)

# Loop through the rows of the DataFrame
for x in df_il.index:
    date_str = str(df_il.loc[x, 'Date'])
    
    # If the date string is longer than 10 characters, truncate it
    if len(date_str) > 10:
        df_il.loc[x, "Date"] = date_str[:10]
    
    # If the date string is in "MonYY" format (e.g., "Jan20"), extract the month and year
    elif len(date_str) == 5:
        month = int(dct_months[date_str[:3]])  # Get the numeric month from the dictionary
        year = int(date_str[3:])  # Get the year part
        
        # Assign the month and year to the respective columns for that row
        df_il.loc[x, 'month'] = month
        df_il.loc[x, 'year'] = year

df_il
```

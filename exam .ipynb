```python
import pandas as pd
```

```python
df_il = pd.read_csv('./data/Israel interest rates.csv')

df_il['Date'] = pd.to_datetime(df_il['DATE'], errors='coerce')
df_il["Year"] = df_il["Date"].dt.year
df_il["Month"] = df_il["Date"].dt.month
df_il["current_rate"] = df_il["IR3TIB01ILM156N"]
df_il["change_from_last_monte"] = df_il["current_rate"].diff()
df_il = df_il[['Date', 'Year', 'Month', 'current_rate', 'change_from_last_monte']]
df_il
```

```python
df_us = pd.read_csv('./data/US interest rates.csv')

df_us['Date'] = pd.to_datetime(df_us['DATE'], errors='coerce')
df_us["Year"] = df_us["Date"].dt.year
df_us["Month"] = df_us["Date"].dt.month
df_us["current_rate"] = df_us["IRLTLT01USM156N"]
df_us["change_from_last_monte"] = df_us["current_rate"].diff()
df_us = df_us[['Date', 'Year', 'Month', 'current_rate', 'change_from_last_monte']]
df_us
```

```python
df_ex = pd.read_csv('./data/USD to ILS exchange rates.csv')

df_ex['Date'] = pd.to_datetime(df_ex['DATE'], errors='coerce')
df_ex["Year"] = df_ex["Date"].dt.year
df_ex["Month"] = df_ex["Date"].dt.month
df_ex["current_rate"] = df_ex["USD"]
df_ex["change_from_last_monte"] = df_ex["current_rate"].diff()
df_ex.drop_duplicates(subset=['Year', 'Month'], inplace=True)
df_ex = df_ex[['Date', 'Year', 'Month', 'current_rate', 'change_from_last_monte']]
df_ex
```
